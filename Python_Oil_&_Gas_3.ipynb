## Completar los valores faltantes: Completar solo aquellas columnas donde los valores nulos son menores


```python
def nullfiller(data,threshhold):
  for col in data.columns:
      frac= data[col].isnull().sum()/len(data)

      if  frac > threshhold:
           data= data.drop(col, axis=1)
      else:
           mean = data[col].mean()
           data [col]= data[col].fillna(mean)
  return data
```

```python
data_nullHandled= nullfiller(data, threshhold= 0.5)
```
